```julia
using Images, ImageShow, FourierTools, TestImages
```

```julia
img = Float32.(testimage("cameraman"))
```

```julia
Gray.(img)
```

```julia
mimg = zeros(size(img))
```

```julia
mimg[:] = img[:] .+ 0.05 .* cos.((1:prod(size(img)))./103f0)
```

`mimg[:]`: represents all elements in the array `mimg`, flattening it into a 1-dimensional view of the array. The `[:]` syntax allows us to modify the entire array in a single operation. When calculating `mimg[:]`, `mimg` must be defined first.

`1:prod(size(img))`: This creates a range of numbers from 1 to `prod(size(img))`, where `prod(size(img))` is the total number of elements (pixels) in `img`. If `img` is a 100x100 `image`, `prod(size(img))` would be $100 \times 100 = 10,000$. This is equivalent to `length(img)`.

`./ 103f0`: Divides each element in the range `1:prod(size(img))` by 103.0 (note the `f0` suffix, which signifies Float32 to ensure floating-point division).

```julia
Gray.(mimg)
```

```julia
fimg = ft(mimg)
```

```julia
function ftshow(a; gamma=0.2)
	m = abs2.(a)
	m = m ./ maximum(m)
	Gray.(m.^gamma)
end
```

```julia
ftshow(fimg)
```

```julia
pos1 = [258, 150]
```

```julia
ctr = size(fimg) ./ 2 .+ 1
```

`ctr` holds the center coordinate of `fimg` in each dimension.

```julia
pos2 = ctr .+ (ctr .- pos1)
```

`ctr .+ (ctr .- pos1)`: Adds this vector to ctr, effectively "mirroring" pos1 across the center. The result pos2 is a new position symmetric to pos1 about the center.

```julia
Δx = 15
```

```julia
fimg[pos1[1] - Δx : pos1[1] + Δx, pos1[2] - Δx : pos1[2] + Δx] .=0
```

```julia
fimg[pos2[1] - Δx : pos2[1] + Δx, pos2[2] - Δx : pos2[2] + Δx] .=0;
```

```julia
ftshow(fimg)
```
